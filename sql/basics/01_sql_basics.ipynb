```python
# Cette fonction nous permet d'écrire les requêtes simplement
def sql(query):
    return pd.read_sql_query(query, conn)

# Test — affiche les clients
sql("SELECT * FROM clients")
```

```python
# Exercice 1 — Affiche tous les clients
sql("SELECT * FROM clients")
```

```python
# Exercice 2 — Affiche uniquement le nom et la ville
sql("SELECT nom, ville FROM clients")
```

```python
# Exercice 3 — Affiche uniquement les clients qui habitent à Paris
sql("SELECT nom FROM clients WHERE ville = 'Paris'")
```

```python
# Affiche les clients de moins de 30 ans, 
# triés du plus jeune au plus âgé
sql("SELECT nom, age FROM clients WHERE age < 30 ORDER BY age ASC")
```

```python
# Question A : combien y a-t-il de clients au total ?
sql("SELECT COUNT (nom) AS nb_clients FROM clients")
```

```python
# Question B : quel est l'âge moyen des clients ?
sql("SELECT ROUND(AVG(age), 1) AS age_moyen FROM clients")
```

```python
# Combien de clients par ville ?
# Trie par nombre décroissant
sql("SELECT ville, COUNT(*) AS nb_clients FROM clients GROUP BY ville ORDER BY nb_clients DESC")
```

```python
# Quelles villes ont PLUS D'UN client ?
sql("""
SELECT ville, COUNT(*) AS nb_clients
FROM clients
GROUP BY ville
HAVING nb_clients > 1
""")
```
